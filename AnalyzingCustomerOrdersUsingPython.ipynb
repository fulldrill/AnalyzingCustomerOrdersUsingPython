```python
#Create a list of customer names
customer_names = ["Matthew", "Sarah", "John", "Emily","Bright", "Olivia", "David", "Sophia", "Michael", "Isabella","CJ"]
print("---Customer List---")
print(customer_names)
```

```python
# Store each customer's order details (customer name, product, price, category) as tuples inside a list
orders = [
    ("Matthew", "Laptop", 1200, "Electronics"),
    ("Sarah", "Smartphone", 800, "Electronics"),
    ("John", "Headphones", 75, "Accessories"),
    ("Bright", "Tshirt", 80, "Clothing"),
    ("Emily", "Camera", 500, "Electronics"),
    ("Michael", "Tablet", 300, "Electronics"),
    ('John', 'Macbook', 1500, 'Electronics'),
    ("Sophia", "Mac mini", 599, "Electronics"),
    ("Olivia", "Laptop", 1100, "Electronics"),
    ("David", "Watch", 99, "Accessories"),
    ("Sophia", "Blouse", 125, "Clothing"),
    ("Isabella", "Camera", 450, "Electronics"),
    ("Bright", "Book", 20, "Books"),
    ("CJ", "Headphones", 39, "Accessories"),
]
print("\n--Customer Orders (as tuples)--")

for o in orders:
    print(o)
 
 
print("\n--Customer Orders (as list)--")
   
for order in orders:
    print(list(order)) #Converts the tuple to a list

```

```python
# Use a dictionary where keys are customer names and values are lists of ordered products
customer_orders = {}
for order in orders:
    customer_name = order[0] #Extract customer name from the tuple [0] is the index for the first element in the tuple
    product = order[1] #Extract product name from the tuple [1] is the index for the second element in the tuple
    if customer_name not in customer_orders: #Check if the customer name is not already a key in the dictionary
        customer_orders[customer_name] = [] #If not, create a new key with an empty list as the value
    customer_orders[customer_name].append(product) #Append the product to the list of products for that customer

print("\n--Customer Orders (as dictionary)--")
for customer, products in customer_orders.items(): #Iterate through the dictionary items, where customer is the key and products is the value (list of products)
    print(f"{customer}: {products}")
```

```python
# #2. Classify products by category
# • Use a dictionary to map each product to its respective category
# • Create a set of unique product categories
# • Display all available product categories
```

```python
# • Use a dictionary to map each product to its respective category
product_categories = {
    "Laptop": "Electronics",
    "Smartphone": "Electronics",
    "Headphones": "Accessories",
    "Camera": "Electronics",
    "Macbook": "Electronics",
    "Tshirt": "Clothing",
    "Blouse": "Clothing",
    "Book": "Books",
    "Watch": "Accessories",
    "Mac mini": "Electronics",
}
print("\n--Product Categories--")
for product, category in product_categories.items():
    print(f"{product}: {category}")
```

```python
# • Create a set of unique product categories
unique_categories = set(product_categories.values()) #Extract the values (categories) from the dictionary and create a set to get unique categories
print("\n--Unique Product Categories--")
print(unique_categories)        
```

```python
# • Display all available product categories
print("\n--Available Product Categories--")
for category in unique_categories:
    print(category)
```

```python
# #3. Analyze customer orders
# • Use a loop to calculate the total amount each customer spends
# • If the total purchase value is above $100, classify the customer as a high-value buyer
# • If it is between $50 and $100, classify the customer as a moderate buyer
# • If it is below $50, classify them as a low-value buyer
```

```python
# • Use a loop to calculate the total amount each customer spends
customer_spending = {}
for order in orders:
    customer_name = order[0] #Extract customer name from the tuple
    price = order[2] #Extract price from the tuple
    if customer_name not in customer_spending: #Check if the customer name is not already a key in the dictionary
        customer_spending[customer_name] = 0 #If not, create a new key with an initial value of 0
    customer_spending[customer_name] += price #Add the price to the total spending for that customer    

print("\n--Customer Spending--")
for customer, total in customer_spending.items():
    print(f"{customer}: ${total}")
```

```python
print("\n--Customer Spending--")
for customer, total in customer_spending.items():
    print(f"{customer}: ${total:.2f}")
```

```python
#If the total purchase value is above $100, classify the customer as a high-value buyer
customer_classification = {}
for customer, total in customer_spending.items():
    if total > 100:
        customer_classification[customer] = "High-value buyer"
    elif total >= 50:
        customer_classification[customer] = "Moderate buyer"
    else:
        customer_classification[customer] = "Low-value buyer"

print("\n--Customer Classification--")
for customer, classification in customer_classification.items():
    print(f"{customer}: {classification}")
```

```python
#Only classify customers as high-value customers if their total purchase value is above $100
high_value_customers = [customer for customer, classification in customer_classification.items() if classification == "High-value buyer"] #List comprehension to create a list of high-value customers
print("\n--High-value Customers--")
print(high_value_customers)
```

```python
moderate_value_customers = [customer for customer, classification in customer_classification.items() if classification == "Moderate buyer"]
print("\n--Moderate-value Customers--")
print(moderate_value_customers)
```

```python
low_value_customers = [customer for customer, classification in customer_classification.items() if classification == "Low-value buyer"]
print("\n--Low-value Customers--")
print(low_value_customers)
```

```python
# #4. Generate business insights
# • Calculate the total revenue per product category and store it in a dictionary
# • Extract unique products from all orders using a set
# • Use a list comprehension to find all customers who purchased electronics
# • Identify the top three highest-spending customers using sorting
```

```python
# • Calculate the total revenue per product category and store it in a dictionary
category_revenue = {} #Initialize an empty dictionary to store revenue per category
for order in orders:  #Iterate through each order in the orders list
    customer, product, price, category = order #Unpack the order tuple into customer, product, price, and category variables
    if category in category_revenue:  #Check if the category is already a key in the category_revenue dictionary
        category_revenue[category] += price #If it is, add the price to the existing revenue for that category
    else: #If it is not, create a new key for that category and set its value to the price of the current order
        category_revenue[category] = price #Initialize the revenue for that category with the price of the current order

print("\n--Total Revenue per Product Category--")
for category, revenue in category_revenue.items(): #Iterate through the category_revenue dictionary items
    print(f"{category}: ${revenue}")
```

```python
# • Extract unique products from all orders using a set
unique_products = set() #Initialize an empty set to store unique products
for order in orders:  #Iterate through each order in the orders list
    customer, product, price, category = order #Unpack the order tuple into customer, product, price, and category variables
    unique_products.add(product) #Add the product to the set of unique products

print("\n--Unique Products--")
for product in unique_products:
    print(product)
```

```python
# • Use a list comprehension to find all customers who purchased electronics
electronics_customers = [customer for customer, product, price, category in orders if category == "Electronics"]
print("\n--Customers who purchased electronics--")
for customer in electronics_customers:
    print(customer)
```

```python
# • Identify the top three highest-spending customers using sorting
customer_spending = {}
for customer, product, price, category in orders: #Iterate through each order in the orders list, unpacking the order tuple into customer, product, price, and category variables
    customer_spending[customer] = customer_spending.get(customer, 0) + price #Add the price to the total spending for that customer

sorted_customers = sorted(customer_spending.items(), key=lambda x: x[1], reverse=True) #Sort the customer_spending dictionary items by total spending in descending order. Lambda function is used as the key to sort by the second element of the tuple (total spending)
top_three_customers = sorted_customers[:3] #Slice the sorted list to get the top three customers

print("\n--Top three highest-spending customers--")
for customer, total_spent in top_three_customers:
    print(f"{customer}: ${total_spent}")
```

```python
# #5. Organize and display data
# • Print a summary of each customer’s total spending and their classification
# • Use set operations to find customers who purchased from multiple categories
# • Identify common customers who bought both electronics and clothing
customer_categories = {}
for customer, product, price, category in orders:
    if customer not in customer_categories:
        customer_categories[customer] = set()
    customer_categories[customer].add(category)

print("\n--Customer spending summary and classifications--")
for customer, total_spent in sorted_customers:
    categories = customer_categories.get(customer, set())
    classification = "High-value" if total_spent > 100 else "Regular"
    print(f"{customer}: ${total_spent} ({classification}) - Categories: {', '.join(categories)}")

# Find customers who purchased from multiple categories
multiple_category_customers = [customer for customer, categories in customer_categories.items() if len(categories) > 1]
print(f"\n--Customers who purchased from multiple categories: {len(multiple_category_customers)}--")
for customer in multiple_category_customers:
    print(customer)

# Identify common customers who bought both electronics and clothing
electronics_customers = {customer for customer, product, price, category in orders if category == "Electronics"}
clothing_customers = {customer for customer, product, price, category in orders if category == "Clothing"}
common_customers = electronics_customers & clothing_customers
print(f"\n--Common customers who bought both electronics and clothing: {len(common_customers)}--")
for customer in common_customers:
    print(customer)
```

```python
# #Customer order processing in python
# • Store customer order data using lists, tuples, and dictionaries

# Create a tuples to store customer orders
customer_orders = (
    ("Matthew", "Laptop", 1200, "Electronics"), #[0] is the index for the first element in the tuple, which is the customer name
    ("Sarah", "Smartphone", 800, "Electronics"),
    ("John", "Headphones", 75, "Accessories"),
    ("Bright", "Tshirt", 80, "Clothing"),
    ("Emily", "Camera", 500, "Accessories"),
    ("Michael", "Tablet", 300, "Electronics"),
    ('John', 'Macbook', 1500, 'Electronics'),
    ("Sophia", "Mac mini", 599, "Electronics"),
    ("Olivia", "Laptop", 1100, "Electronics"),
    ("David", "Watch", 99, "Accessories"),
    ("Sophia", "Blouse", 125, "Clothing"),
    ("Isabella", "Camera", 450, "Electronics"),
    ("Bright", "Book", 20, "Books"),
    ("CJ", "Headphones", 39, "Accessories"),
)
print("---Customer Orders Tuple---")
for order in customer_orders:
    print(order)

#Example of retrieving customer orders
print("\n--Customer Orders for Sophia--")
for order in customer_orders:
    if order[0] == "Sophia": #Check if the first element of the order tuple (customer name) is "Sophia"
        print(order)

#Example of modifying a customer's order
#Note: Tuples are immutable, so we cannot modify an existing tuple. Instead, we
# create a new tuple with the updated order details.
updated_orders = []
for order in customer_orders:
    if order[0] == "Sophia": #Check if the first element of the order tuple (customer name) is "Sophia"
        updated_orders.append(("Sophia", "Mac mini", 599, "Electronics")) #If it is, append a new tuple with the updated details to the updated_orders list
    else:
        updated_orders.append(order) #If it is not, append the original order tuple to the updated_orders list
customer_orders = tuple(updated_orders) #Convert the updated_orders list back to a tuple
print("\n--Updated Customer Orders Tuple--")
for order in customer_orders:
    print(order)
```

```python
# Classification and analysis
# • Use loops to categorize customers based on their total spending
customer_spending = {}

for customer_name, _, price, _ in customer_orders:
    customer_spending[customer_name] = customer_spending.get(customer_name, 0) + float(price)

print("\n--Customer Spending--")
for customer, total in customer_spending.items():
    print(f"{customer}: ${total:.2f}")

#Use set operations to find common and unique products across different categories
electronics_products = {product for _, product, _, category in customer_orders if category == "Electronics"}
clothing_products = {product for _, product, _, category in customer_orders if category == "Clothing"}
common_products = electronics_products & clothing_products
unique_electronics = electronics_products - clothing_products
print("\n--Common Products in Electronics and Clothing--")
if not common_products:
    print("There are no common products between Electronics and Clothing categories.")
else:
    for product in common_products:
        print(product)

electronics_products = {product for _, product, _, category in customer_orders if category == "Electronics"}
accessories_products = {product for _, product, _, category in customer_orders if category == "Accessories"}
common_products = electronics_products & accessories_products
print("\n--Common Products in Electronics and Accessories--")
if not common_products:
    print("There are no common products between Electronics and Accessories categories.")
else:
    for product in common_products:
        print(product)
```

```python
# Insight generation
# • Extract the high-value customers 
high_value_customers = [customer for customer, total in customer_spending.items() if total > 100]
print("\n--High-value Customers--")
for customer in high_value_customers:
    print(customer)

#Extract the most frequently purchased products
most_frequent_purchased_products = {}
for _, product, _, _ in customer_orders:  #Iterate through each order in the customer_orders tuple, unpacking the order tuple into customer_name, product, price, and category variables
    most_frequent_purchased_products[product] = most_frequent_purchased_products.get(product, 0) + 1 #Count the frequency of each product by adding 1 to the count for that product in the most_frequent_purchased_products dictionary
sorted_products = sorted(most_frequent_purchased_products.items(), key=lambda x: x[1], reverse=True) #Sort the products by frequency in descending order using a lambda function as the key
print("\n--Most Frequently Purchased Products--")
for product, frequency in sorted_products:
    print(f"{product}: {frequency} purchases")


# • Identify trends based on category-wise sales
category_sales = {}
for _, _, price, category in customer_orders: #Iterate through each order in the customer_orders tuple, unpacking the order tuple into customer_name, product, price, and category variables
    category_sales[category] = category_sales.get(category, 0) + price #Calculate total sales for each category by adding the price to the existing total for that category in the category_sales dictionary
sorted_categories = sorted(category_sales.items(), key=lambda x: x[1], reverse=True) #Sort the categories by total sales in descending order using a lambda function as the key
print("\n--Category-wise Sales Trends--")
for category, total_sales in sorted_categories:
    print(f"{category}: ${total_sales:.2f}")    
```
